```python
#Plotting the graph for having values without scaling
#Sorting the having dataset on the basis of stability and unstability
# First 3620 rows are Stable and 6380 are unstable grid points using above line
data1=data.sort_values(['stab'], axis=0, 
                 ascending=True, inplace=False) 

data1['New_ID'] = range(1, 1+len(data1))
data1.set_index("New_ID", inplace = True)
plt.scatter(data1.iloc[:300,0],data1.iloc[:300,1],color='red',marker='o',label='Stable')
plt.scatter(data1.iloc[-300:,0],data1.iloc[-300:,1],color='blue',marker='*',label='UN Stable')
plt.plot()
plt.show()
#SCATTER PLOT FOR STABLE AND UNSTABLE GRIDS BASED ON FIRST 2 FEATURE
```
